```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('BlackFriday.csv')
df.isna().sum()
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
lst=[]
for i in range(len(df)):
    lst.append(df['Gender'][i]+'_'+ str(df['Marital_Status'][i]))

df['MaritalGender']=lst
```

```python
df.head()
```

```python
sns.countplot(x='MaritalGender',data=df)
```

```python
sns.countplot(x='MaritalGender',data=df, hue = 'Age')
```

```python
sns.countplot(hue='MaritalGender',data=df, x= 'Product_Category_1')
```

```python
sns.countplot(hue='MaritalGender',data=df, x= 'Stay_In_Current_City_Years')
```
