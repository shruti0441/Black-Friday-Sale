```python
import pandas as pd
```

```python
df= pd.read_csv('BlackFriday.csv')
df.isna().sum()
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
for column in df.columns:
    print(column,";", df[column].nunique())
```

```python
df.groupby('Gender').size().plot(kind='pie',autopct='%.1f', figsize=(6,6),title='Gender Ratio')
```

```python
sum=pd.DataFrame(df.groupby('Gender')['Purchase'].sum())
```

```python
sum
```

```python
sum['Purchase'].plot(kind= 'pie', autopct='%.1f')
```

```python
df.groupby('Gender')['Purchase'].mean().plot(kind='pie',autopct='%.1f')
```
