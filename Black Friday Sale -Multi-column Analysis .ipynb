```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('BlackFriday.csv')
df.isna().sum()
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age')['Purchase'].size().plot(kind='bar',figsize=(10,5),title = 'Production Distribution by Age')
```

```python
sns.set(rc= {'figure.figsize': (10,6)})
sns.countplot(x= 'Age', hue= 'Gender' , data = df)
```

```python
sns.countplot(data=df, x= 'Gender',hue= 'Age')
```

```python
sns.set(rc= {'figure.figsize':(5,5)})
sns.countplot(x='Gender', hue='Marital_Status',data= df)
```

```python
sns.countplot(data=df, x='Marital_Status', hue= 'Gender')
```

```python
sns.countplot(x=df['City_Category'])
```

```python
sns.countplot(data=df ,x= 'City_Category', hue= 'Gender')
```

```python
sns.countplot(data=df ,hue= 'City_Category',x= 'Gender')
```

```python
sns.countplot(data=df ,x= 'City_Category', hue= 'Marital_Status')
```

```python
sns.countplot(data=df ,hue= 'City_Category', x= 'Marital_Status')
```

```python
sns.countplot(data=df ,x= 'City_Category', hue= 'Age')
```

```python
df.groupby('City_Category').size().plot(kind='pie',autopct='%.2f',title= 'Purchase ditribution among cities')
```

```python
df.groupby('City_Category')['Purchase'].sum().plot(kind='pie',autopct= '%.1f')
```

```python
df.groupby('City_Category')['Purchase'].mean().plot(kind='pie',autopct='%.2f',title= 'Purchase ditribution among cities')
```
